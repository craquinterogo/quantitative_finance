# Geometric Brownian Motion

This is one way to model price dynamics ....

$dS = \mu S dt + \sigma S dW$

where $W_t$ is the Brownian motion ....

$S(t) = S_0 e^{(\mu + \frac{1}{2}\sigma^2)t + \sigma W_t}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def geometric_brownian_motion(S0, mu, sigma, T, n_sim):
    """
    Function to simulate Geometric Bronian Motion Prices
    """
    dt = 1 / T
    Z = np.random.standard_normal(size=(n_sim,T))
    dw = Z * np.sqrt(dt)
    W = np.cumsum(dw, axis=1)
    
    St = S0 * np.exp((mu-0.5*sigma**2)*T + sigma*W)
    St = np.hstack((np.transpose([[S0]*n_sim]), St))
    return St
```

```python
St = geometric_brownian_motion(S0=100, mu=0.2, sigma=0.7, T=10, n_sim=1000)
```

```python
plt.figure(figsize=(18,10))
plt.plot(St.T)
plt.grid()
plt.show()
```
